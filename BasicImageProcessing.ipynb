```python
from IPython import get_ipython
from IPython.display import display
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow from google.colab.patches


```

```python
# Reading an image
img = cv2.imread("/content/fruits")
print(type(img))
```

```python
img.shape
```

```python
# Display the image

cv2_imshow(img)

```

```python
# Converting RGB to Grey-scale

img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_grey)
```

```python
print(img_grey.shape)
```

```python
# Playing with RGB channels
img[:,:,0] = 0                # Blue channel removed as 0 : Blue, 1: Green and 2 : Red
cv2_imshow(img)
```

```python
img[:,:,1] = 0                # Green channel removed as 0 : Blue, 1: Green and 2 : Red
cv2_imshow(img)
```

```python
img[:,:,2] = 0                # Red channel removed as 0 : Blue, 1: Green and 2 : Red
cv2_imshow(img)
```

```python
img2 = cv2.imread("/content/fruits")
```

```python
# Image resize

img_resize = cv2.resize(img2, (256,256))
cv2_imshow(img_resize)
```

```python
# Flipping an Image

img_flip1 = cv2.flip(img2, 0)                         # Flip vertically
img_flip2 = cv2.flip(img2, 1)                         # Flip Horizontally
img_flip3 = cv2.flip(img2, -1)                         # Combined effect of both
cv2_imshow(img_flip1)
cv2_imshow(img_flip2)
cv2_imshow(img_flip3)
```

```python
# Cropping an Image

img_crop = img2[100:300, 200:500]                     # Height axis : 100 to 300 and Width axis : 200 to 500
cv2_imshow(img_crop)
```

```python
# Saving an image

cv2.imwrite("fruits_small.png",img_crop)                      # True defines that image has been successfully saved on your local folder
```

```python
import numpy as np
```

```python
# Drawing an image

img = np.zeros((512,512,3))
cv2_imshow(img)
```

For drawing rectangle we have built-in function cv2.rectangle() where we pass following parameters :

1) img_name : Image on which rectangle will be created

2) pt1 = (x,y) : Starting point of rectangle

3) pt2 = (m,n) : Ending point of rectangle

4) color = ((x,y,z),q) : Colour of the rectangle where x,y,z denotes b,g,r respectively and q denotes thickness

```python
# Drawing Rectangle

cv2.rectangle(img, pt1 = (100,100), pt2 = (300,300), color = (255,0,0), thickness = 3)
cv2_imshow(img)
cv2.rectangle(img, pt1 = (100,100), pt2 = (300,300), color = (255,0,0), thickness = -1)
cv2_imshow(img)
```

```python
# Drawing circle

cv2.circle(img, center = (100,400), radius = 50, color = (0,0,255), thickness = 3)
cv2_imshow(img)
```

```python
# Drawing line

cv2.line(img, pt1 = (0,0), pt2 = (512,512), color = (0,255,0), thickness = 2)
cv2_imshow(img)
```

```python
# Adding text

cv2.putText(img,org = (100,100), fontFace = cv2.FONT_ITALIC, fontScale = 4, color = (0,255,255), thickness = 2, lineType = cv2.LINE_AA, text = "Hello")
cv2_imshow(img)
```
